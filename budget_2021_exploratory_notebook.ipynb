```python
import pandas as pd
import json
import requests


budget_2021 = pd.read_csv("https://raw.githubusercontent.com/swatland610/data_viz_challenge/main/data/2021_chicago_budget_appropriations.csv")
budget_2021.info()
```

```python
# group budgets by department
departments = budget_2021.groupby('DEPARTMENT DESCRIPTION').agg({'2021 ORDINANCE (AMOUNT $)':'sum'})

departments.sort_values(by="2021 ORDINANCE (AMOUNT $)", ascending=False)
```

```python
# subset all from DFSS
dfss_budget_line_items = budget_2021[budget_2021['DEPARTMENT DESCRIPTION']=='CPD']
```

```python
dfss_budget_line_items.groupby("APPROPRIATION AUTHORITY DESCRIPTION")\
    .agg({'2021 ORDINANCE (AMOUNT $)':'sum'})\
    .sort_values(by="2021 ORDINANCE (AMOUNT $)", ascending=False)
```
